### 메타데이터 - 행정동 제외한 나머지

```python
wow = pd.read_excel('meta_data/2024_빅콘테스트_데이터분석분야_챔피언부문_데이터정의서(SKT).xlsx')
wow
```

```python
why = wow.iloc[28:, :2].reset_index(drop = True)
# why.set_index('테이블정의서', inplace=True)
why.columns = ['코드' ,'종류']
why = pd.concat([pd.DataFrame({'코드종류' : 'purpose'}, index=range(6)), why], axis=1)

why
```

```python
modal = wow.iloc[18:26, :2].reset_index(drop=True)
# modal.set_index('테이블정의서', inplace=True)
modal.columns = ['코드', '종류']
modal = pd.concat([pd.DataFrame({'코드종류' : 'modal'}, index=range(len(modal))), modal], axis=1)
modal
```

```python
gender = pd.DataFrame(
    [
        ['gender', 0, '남자'],
        ['gender', 1, '여자']
    ],
    columns = modal.columns
)

gender
```

```python
age = pd.DataFrame(
    [
        ['age', i, f'{i*10} ~ {i*10 +9}세' if i < 8 else '80대 이상'] for i in range(9)
    ],
    columns=modal.columns
)

age
```

```python
decoder = pd.concat([gender, age, modal, why])
decoder.to_csv('feature_scaled_data/decoder.csv', index=False)
```
