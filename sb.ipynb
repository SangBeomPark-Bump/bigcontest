# 10/7, 10/14, 9/23 여의도 데이터

```python
import pandas as pd
```

```python
hdong2 = pd.read_csv('feature_scaled_data/hdong_2.csv')
hdong2.set_index('행정동코드', inplace=True)
```

```python
oct_1007 = pd.read_csv('datas/od_20231001_15/od_20231007_1.csv')
oct_1007
```

```python
oct_1007_orhcd = oct_1007.groupby(['origin_hdong_cd'])['od_cnts'].sum().sort_values(ascending=False)
oct_1007_orhcd
```

```python
hdong2.loc[oct_1007_orhcd.head().index,['주소']]
```

```python
del oct_1007
```
